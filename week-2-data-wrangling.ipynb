```python
###STEP 2:STRUCTURING

#converting 'date_added' to datetime
df['date_added']=pd.to_datetime(df['date_added'],format='mixed')
```

```python
#Separating 'duration' into numeric value and unit (e.g., '90 min' → 90, 'min')
df[['duration_value', 'duration_unit']] = df['duration'].str.extract(r'(\d+)\s*(\w+)')
```

```python
#Converting duration_value to numeric
df['duration_value'] = pd.to_numeric(df['duration_value'])
```

```python
#Viewing resulting columns
print(df[['duration_value', 'duration_unit']])
```

```python
###STEP 3:CLEANING

#Checking for duplicate rows
print("Duplicate rows before:", df.duplicated().sum())
```

```python
#Drop duplicate rows if any
df = df.drop_duplicates()
```

```python
#Drop description column because it will not be used
df = df.drop(columns=['description'])
```

```python
#Impute Director values by using relationship between cast and director
#List of Director-Cast pairs and the number of times they appear
df['dir_cast']=df['director']+"---"+df['cast'] #Build director–cast pair
counts=df['dir_cast'].value_counts() #Counts unique values
filtered_counts=counts[counts>=2] #Checks if is repeated 2 or more times
filtered_values=filtered_counts.index #Keeps only pairs that repeat at least 2 times
list_dir_cast=list(filtered_values) #converts to list
#LOOP:building dictionary {cast: director}
dict_dircast={} 
for i in list_dir_cast: #loop goes through each "director---cast" pair
    director,cast=i.split('---') 
    dict_dircast[cast]=director #add mapping {cast: director}
```

```python
#Filling missing directors using dictionary
df.loc[df['director'].isna(),'director']=df.loc[df['director'].isna(),'cast'].map(dict_dircast)
```

```python
#Assigning Not Given to all other director fields
df.loc[df['director'].isna(),'director']='Not Given'
```

```python
#Using directors to fill missing countries
directors=df['director']
countries=df['country']
```

```python
#Pairing each director with their country use zip() to get an iterator of tuple
pairs=zip(directors,countries)
```

```python
#Converting the list of tuples into a dictionary
dir_cntry = dict(list(pairs))
```

```python
#Director matched to Country values used to fill in null country values
df.loc[df['country'].isna(),'country']=df.loc[df['country'].isna(),'director'].map(dir_cntry)
```

```python
#Assigning Not Given to all other country fields
df.loc[df['country'].isna(),'country'] = 'Not Given'
```

```python
#Dropping other row records that are null
df.drop(df[df['date_added'].isna()].index,axis=0,inplace=True)
df.drop(df[df['rating'].isna()].index,axis=0,inplace=True)
df.drop(df[df['duration'].isna()].index,axis=0,inplace=True)
```

```python
###STEP 4:ERRORS CHECKING

#Checking if there are any added_dates that come before release_year
import datetime as dt
sum(df['date_added'].dt.year<df['release_year']) #Counting invalid rows 
df.iloc[[1551,1696,2920,3168]][['title','date_added','release_year']] #Sample rows to inspect manually with title for easy clarity
```

```python
df=df.loc[~(df['date_added'].dt.year<df['release_year'])] #Droping everything else expect rows where date_added year is greater than or equal to release_year
sum(df['date_added'].dt.year<df['release_year'])
```

```python
###STEP 5:VALIDATION

df.drop(columns=['dir_cast'], inplace=True) #Droping helper columns
```

```python
#checking that no important fields are still missing
print(df.isnull().sum())#checking which columns have missing data and how many missing values there are
df = df.dropna(subset=['title','release_year','date_added']) #droping rows where important fields are missing
```
